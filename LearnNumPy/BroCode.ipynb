# Making numpy arrays

```python
import numpy as np
```

```python
arr = np.array([[7, 8, 9], [11, 45, 66]])
print(arr)

arr *= 2
print(arr)
```

# Multidimentional arrays

```python
arr2 = np.array(
    [
        [[3, 5, 6], ["h", "g", 7], [1, 5, 4]],
        [[7, 5, 8], [5, 2, 4], [4, 5, 9]],
        [["y", 5, 6], [4, "i", "6"], [1, 4, 2]],
    ]
)

print(arr2)
 
print(arr2.shape)

print(arr2[1, 2, 2])
# or
print(arr2[1][2][2])
```

# Slicing 

```python
arr3 = np.array(
    [
        [3, 5, 6],
        [7, 5, 8],
        ["y", 5, 6]
    ]
)

print(arr3[2::1][0::2])
```
